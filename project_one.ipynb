```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv(r'C:\Users\USER\Desktop\data science projects\retail_env\Global+Electronics+Retailer (1)\Customers.csv', encoding='ISO-8859-1')

# Display the first 5 rows of the DataFrame
df.head()
```

```python
##1. Datetime Transformation ##
```

```python
df['Birthday'] = pd.to_datetime(df['Birthday'])
df.head()
```

```python
from datetime import datetime

df['Age'] = datetime.now().year - df['Birthday'].dt.year
df.head()
```

```python
df['Birth Month'] = df['Birthday'].dt.month
df['Birth Year'] = df['Birthday'].dt.year
df.head()
```

```python
### Geographical Transformation ###
```

```python
df['Full Address'] = df['City'] + ', ' + df['State'] + ', ' + df['Zip Code'].astype(str)
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check for duplicate rows in the DataFrame
duplicate_rows = df[df.duplicated()]

# Display duplicate rows
print("Duplicate rows in the DataFrame:")
print(duplicate_rows)
```

```python
column_name = 'Name'

# Check for duplicate values in a specific column
duplicate_in_column = df[df[column_name].duplicated()].sort_values('Name')

# Display duplicate values in the specified column
print(f"Duplicate values in the '{column_name}' column:")
print(duplicate_in_column)



```

```python
# DataFrame 'df' is already defined
column_name = 'Name'

# Check for duplicate values in the specified column
# 'keep=False' ensures all duplicates are flagged, not just the subsequent ones
duplicate_in_column = df[df[column_name].duplicated(keep=False)].sort_values(by=column_name)

# Display the number of duplicate entries
print(f"Total number of duplicate values in the '{column_name}' column: {duplicate_in_column.shape[0]}")

# Display the duplicate entries
print("Duplicate values in the specified column:")
print(duplicate_in_column)
```
